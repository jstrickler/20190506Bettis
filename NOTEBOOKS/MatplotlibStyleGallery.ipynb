# Matplotlib/Seaborn Style gallery
Matplotlib and seaborn have many builtin predefined styles. This notebook plots the same data using each different style.

## Required imports

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating the data to plot
We'll just create some cosine data to plot

```python
x_data = np.arange(0.0, 5.0, 0.1)
y_data = np.cos(x_data)
```

## Plot all the styles!
Loop through all available styles and plot the data.

```python
mpl.rcParams['figure.max_open_warning'] = 25  # allow more than 20 figures without a warning
for i, style_name in enumerate(mpl.style.available, 1):
    if style_name.startswith('_'):
        continue  # skip "private" styles
    plt.figure()
    mpl.style.use(style_name)
    plt.title(style_name)
    plt.plot(x_data, y_data)
```
